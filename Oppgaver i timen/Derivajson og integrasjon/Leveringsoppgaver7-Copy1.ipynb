### 7.9  *feil oppgave

Definisjonen av integralet er at det er arealet under grafen. For å finne integralet kan vi derfor bruke rektangler til å finne arealet. Dette gjør vi ved begrense bredden til rektanglene, og høyden er $f(x_i)$ der $x_i$ er x-kordinatet. Hvis vi da regner ut arealet til n-antall rektangler, mellom to punkter a og b, får vi formelene:
$\Delta x$ = $\frac{b-a}{n}$. 
$$\int_a^b f(x) dx = \Delta x *f(x_a)+\Delta x *f(x_a + \Delta x)... \Delta x *f(x_b)  $$

### 7.3 :


```python
from pylab import *
e = 2.718 #Dette er ca. hva eulertallet er

def N(t): 
    return (0.2*e**t)+90

def derivert(N, t, dt):
    return (N(t + dt) - N(t))/dt

antall = []
endringen = []
tid = []    #år

for i in range(11):
    tid.append(i)
    antall.append(N(i))
    endringen.append(derivert(N, i, 10**-8))
    
plot(tid, antall)
plot(tid, endringen)
legend(["antall planter", "endringen av antall planter"])
show()
```

```python
from pylab import *

tid =[]
temp = []
endring =[]

def T(t):
    return 70*exp(-0.065*t)

def der(f,x,dx =1E-8):
    fder = (f(x+dx)-f(x))/dx
    return fder

for i in range(0,6001):
    x = i/1000
    tid.append(x)
    endring.append(der(T,x,))
    temp.append(T(x))
    
plot(tid, temp)
plot(tid, endring)
legend(["temperatur", "endring i temperatur"])
show()

print("Ved element 42 synker temperaturen med ", endring[42])
```

### Oppgave 7.12

```python
from pylab import *

data = loadtxt("posisjon.txt", skiprows = 1)
tid = data[:,0]
posisjon = data[:,1]
fart = []
aks = []

for i in range(len(tid)-1):
    dy = posisjon[i+1]-posisjon[i]
    dx = tid[i+1]-tid[i]
    der = dy/dx
    fart.append(der)

for i in range(len(tid)-2):
    dy = fart[i+1]-fart[i]
    dx = tid[i+1]-tid[i]
    der = dy/dx
    aks.append(der)
    
plot(tid, posisjon)
plot(tid[:-1], fart)
plot(tid[:-2], aks)

legend(["Posisjon","Fart","Akselerasjon"])
xlabel("Tid (s)")
show()
```

```python
from pylab import *

def f(x):
    return 2*x + 1
def g(x):
    return x**2-4*x+5
def h(x):
    return exp(x)
    

def integralet(f, a, b, N):
    A = 0
    dx = (b - a)/N
    for i in range(N):
        A = A + f(a + i*dx)*dx
    return A

print(integralet(f, 2, 5, 3))
print(integralet(g,2,5,3))
print(integralet(h,2,5,3))
```
